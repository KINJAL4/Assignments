```python
import scipy.stats as s
```

```python
import numpy as np
```

```python
import seaborn as sns
```

# for sample size(1000,2)

```python
x=np.random.exponential(scale=1.0,size=(1000,2))
```

```python
mvue_values =np.reciprocal(np.mean(x,axis=1))
samples_mean = np.mean(x,axis=1)
```

```python
best_estimate=np.mean(mvue_values)
best_mean=np.mean(samples_mean)
```

```python
best_estimate
```

```python
best_mean
```

```python
log_like_max_value =np.sum(s.expon.logpdf(x,scale=best_estimate))
```

```python
log_like_max_value
```

```python
np.var(mvue_values)
```

```python
np.var(samples_mean)
```

```python
sns.distplot(mvue_values,bins=10,kde=True,fit=s.expon)
```

```python
sns.distplot(samples_mean,bins=10,kde=True,fit=s.expon)
```
