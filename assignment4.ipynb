```python
def is_even(number):
    if number % 2==0:
        return True
    else:
        return False
    return is_even(number)
```

```python
def is_prime(number,divisor=2):
    if number==1 or number==2:
        return True
    if number % divisor == 0:
        return False
    if divisor==(number-1):
        return True
        divisor=divisor+1
    return is_prime(number,divisor+1)  
```
